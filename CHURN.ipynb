 ●id: Unique identifier for each customer.
 ●channel_sales: Represents the sales channel through which the customer was acquired.
 ● cons_12m: Represents the electricity consumption in the last 12 months.
● cons_gas_12m: Represents the gas consumption in the last 12 months.
● cons_last_month: Represents the electricity consumption in the last month.
● date_activ: Date when the contract was activated.
● date_end: Date when the contract is scheduled to end.
● date_modif_prod: Date of the last modification of the product.
● date_renewal: Date when the contract is scheduled to be renewed.
● forecast_cons_12m: Forecasted electricity consumption for the next 12 months.
● has_gas: Indicates whether the customer has a gas supply (True/False).
● imp_cons: Represents important consumption.
● margin_gross_pow_ele: Gross margin on power for electricity.
● margin_net_pow_ele: Net margin on power for electricity.
● nb_prod_act: Number of active products.
● net_margin: Net margin for the customer.
 ●num_years_antig: Number of years as a customer (antiquity).
 ● origin_up: code of the electricity campaign the customer first subscribed to /Origin of the contract.
● pow_max: Maximum power.
●churn: Indicates whether the customer has churned (1) or not (0).









price_data.csv
● id = client company identifier
● price_date = reference date
● price_off_peak_var = price of energy for the 1st period (off peak)
● price_peak_var = price of energy for the 2nd period (peak)
● price_mid_peak_var = price of energy for the 3rd period (mid peak)
● price_off_peak_fix = price of power for the 1st period (off peak)
● price_peak_fix = price of power for the 2nd period (peak)
● price_mid_peak_fix = price of power for the 3rd period (mid peak)
